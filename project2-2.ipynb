```python
""" 
Objective: Implement core features of a music player, including managing a list of favorite songs and tracking 
the playback history, using Python built-in list as a container and a custom stack data structure.

Core Data Structures to Implement and Use
1. Song Class
2. Favorite Songs Container (Python list)
3. Custom Stack Class
"""
class Song():
    def __init__(self,title=str,artist=str):
        self.t = title
        self.a = artist
    def __eq__(self,other):
        return self.t.lower() == other.t.lower() and self.a.lower() == other.a.lower()
    def __str__(self):
        return f'Title is: {self.t} and artist is: {self.a}'

class Stack():
    def __init__(self,L=[]):
        self.l = L
        self.size=0
    def push(self,item):
        self.l.append(item)
        self.size+=1
    def pop(self):
        if len(self.l)==0:
            print('Stack is empty')
            return None
        self.size-=1
        return self.l.pop()
    def peek(self):
        if len(self.l)==0:
            print('Stack is empty')
            return None
        print(self.l[-1])
        return
    def is_empty(self):
        return self.size == 0

fav_list=[]
playback_history_stack = Stack()

def menu(user):
    
    if user == 1:
        title, artist = input('\nEnter the song title: '), input('Enter the artist: ')
        song = Song(title, artist)
        for c in fav_list:
            if song == c:
                print(f'{song} already exist')
                return
        fav_list.append(song)
        
    elif user ==2:
        if len(fav_list)==0:
            print('List is empty')
            return 
        title, artist = input('\nEnter the song title: '), input('Enter the artist: ')
        temp = Song(title, artist)
        if temp not in fav_list:
            print('Song not found')
        else:
            fav_list.remove(temp)
        
    elif user==3:
        title, artist = input('\nEnter the song title: '), input('Enter the artist: ')
        song = Song(title, artist)
        playback_history_stack.push(song)
        print(f'\nNow playing: {title} by {artist}')

    elif user==4:
        if playback_history_stack.is_empty()==True or playback_history_stack.size == 1:
            print('\nNo previous songs in history.')
        playback_history_stack.pop()
        print(f'\nNow Playing: ',end='')
        playback_history_stack.peek()
            
    elif user==5:
        if len(fav_list)==0:
            print('\nList is empty')
            return 
        print('')
        for i,s in enumerate(fav_list, start=1):
            print(f'{i}. {s}')
    
    elif user==6:
        if playback_history_stack.is_empty():
             print('\nPlayback history is empty.')
             return
        temp = []
        print('\nPlayback History')
        while playback_history_stack.size != 0 :
            song = playback_history_stack.pop() 
            temp.append(song)
        
        for i, s in enumerate(temp, start=1):
            print(f'{i}. {s}')
        for s in reversed(temp):
            playback_history_stack.push(s)
        
    elif user==7: 
        print('\nExiting the program.')
        return False
    else:
        print('\nRechoose option 1-7')

if __name__ =='__main__':
    user=0
    while user != 7:
        user = int(input('\n– 1. Add Song to Favorites\n– 2. Remove Song from Favorites\n– 3. Play Song\n– 4. Go Back to Previous Song\n– 5. View Favorites\n– 6. View Playback History\n– 7. Exit\n'))
        menu(user)
```
